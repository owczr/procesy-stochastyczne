```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
SAMPLES = 100
```

```python
X = np.linspace(0, 1, 100)
```

```python
y_norm = np.random.normal(loc=0, scale=1, size=SAMPLES)
y_poisson = np.random.poisson(lam=1, size=SAMPLES)
y_log = np.random.lognormal(mean=0, sigma=1, size=SAMPLES)
y_exp = np.random.exponential(scale=1, size=SAMPLES)
```

```python
fig = plt.figure(figsize=(15, 5))
plt.plot(X, y_norm, label="normal", c="b")
plt.plot(X, y_poisson, label="poisson", c="orange")
plt.plot(X, y_log, label="log", c="g")
plt.plot(X, y_exp, label="exponential", c="r")
plt.legend()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes[0, 0].hist(y_norm, color="blue")
axes[0, 0].set_title("Normal")
axes[0, 1].hist(y_poisson, color="orange")
axes[0, 1].set_title("Poisson")
axes[1, 0].hist(y_log, color="green")
axes[1, 0].set_title("Log")
axes[1, 1].hist(y_exp, color="red")
axes[1, 1].set_title("Exponential")
```

```python
def stats(values, name):
    print(name)
    print(f"Min:    {np.min(values)}")
    print(f"Max:    {np.max(values)}")
    print(f"Mean:   {np.mean(values):.2f}")
    print(f"Median: {np.median(values)}", end="\n\n")
```

```python
stats(y_norm, "Normal")
stats(y_poisson, "Poisson")
stats(y_log, "Log")
stats(y_exp, "Exponential")
```
